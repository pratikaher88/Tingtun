```python
import json
import pandas as pd
import numpy as np
file = 'dexp2'
with open(file) as train_file:
    dict_train = json.load(train_file)
```

```python
# print(dict_train)

dict_train = pd.io.json.json_normalize(dict_train)

train = pd.DataFrame.from_dict(dict_train)

train
```

```python
train[['barrier_report.what_went_wrong','webpage_reported']]
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 20]
```

```python
train[['webpage_reported']].apply(pd.value_counts).plot(kind='bar', subplots=True)
plt.show()
```

```python
train[['barrier_report.what_went_wrong']].apply(pd.value_counts).plot(kind='bar', subplots=True)
plt.show()
```

```python
df=train[['barrier_report.what_went_wrong','webpage_reported']]
```

```python
df=df[df['barrier_report.what_went_wrong'].str.len() > 0]
df=df[df['webpage_reported'].str.len()>0]
```

```python
df['barrier_report.what_went_wrong'] = df['barrier_report.what_went_wrong'].astype(str).str.lower()
```

```python
df['Counts']=df.groupby(['barrier_report.what_went_wrong'])['webpage_reported'].transform('count')
```

```python
df=df.sort_values('Counts',ascending=False)
```

```python
df=df.drop_duplicates().reset_index(drop=True)
```

```python
df=df.sort_values('webpage_reported')
```

```python
df.to_json('report.json', orient='records', lines=True)
```
